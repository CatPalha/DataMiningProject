## Removing Outliers ##

```python
import pandas as pd
import seaborn as sb
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('A2Z_insurance.csv')
df.head()
```

```python
df.columns.values
```

**I'm gonna replace the NaN values with zero because I wanna check if the sum of the premims is greater that the salary of each costumer**

```python
df = df.fillna(0)

df.isnull().sum()
```

```python
df['premiums_sum'] = df['Premiums in LOB: Motor'] +  df['Premiums in LOB: Household'] + df['Premiums in LOB: Health'] + df['Premiums in LOB:  Life'] + df['Premiums in LOB: Work Compensations']
```

```python
df[df['premiums_sum'] > df['Gross Monthly Salary']]
#621 costumers 
```

**Box plot on each continuos variable to spot the outliers**

```python
sb.boxplot(data=df[['Gross Monthly Salary', 'Customer Monetary Value', 'Premiums in LOB: Motor', 'Premiums in LOB: Household', 'Premiums in LOB: Health',
                    'Premiums in LOB:  Life', 'Premiums in LOB: Work Compensations']], orient = 'h', whis = 3.5)

plt.show()
```

```python
#Let's normalize the data

df_to_norm = df[['Gross Monthly Salary', 'Customer Monetary Value', 'Premiums in LOB: Motor', 'Premiums in LOB: Household', 'Premiums in LOB: Health',
    'Premiums in LOB:  Life', 'Premiums in LOB: Work Compensations']]

def z_score(df): 
    
    return (df - df.mean()) / df.std()

df_to_norm.apply(z_score)
```

```python
#Plot it again

sb.boxplot(data = df_to_norm, orient = 'h', whis = 3.5)
```
