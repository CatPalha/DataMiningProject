The data appears to be skewed so we'll replace the missing values with the median

```python
median = df['Gross Monthly Salary'].median()
df_median_replace = df.copy(True)
df_median_replace['Gross Monthly Salary'] = df_median_replace['Gross Monthly Salary'].fillna(median)
descr_median = df_median_replace.describe()
descr_median
```

```python
descr = df.describe()
descr
```

```python
mean = df['Gross Monthly Salary'].mean()
df_mean_replace = df.copy(True)
df_mean_replace['Gross Monthly Salary'] = df_mean_replace['Gross Monthly Salary'].fillna(mean)
descr_mean = df_mean_replace.describe()
descr_mean
```

```python
edu = df['Educational Degree'].unique()
edu
```

```python
edu_counts = df.groupby(['Educational Degree'])['Educational Degree'].count()
edu_counts
```

```python
fr = []
total = edu_counts.values.sum()
for edu in edu_counts.values:
    freq = edu/total
    fr.append(freq)
fr
```

```python
na_total = df['Educational Degree'].isna().sum()
print(na_total)
fr_na = []
for freq in fr:
    freq_na = na_total * freq
    fr_na.append(freq_na)
fr_na
```
